```python
# --- Step 1: Import the dataset ---
import pandas as pd
import numpy as np

# Load Excel file (make sure it's in the same folder)
file_path = "MLL_4.xlsx"
df = pd.read_excel(file_path)

# Show first few rows and info
print("âœ… Dataset Loaded Successfully")
print(df.head())
print(df.info())
```

```python
# --- Step 2: Randomize the dataset ---
df.columns = df.columns.astype(str)  # Ensure all column names are strings
df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle rows
print("âœ… Dataset Randomized")
```

```python
df
```

```python
# --- Step 3: Normalize (MinMaxScaler from -1 to 1) ---
from sklearn.preprocessing import MinMaxScaler

target_col = 'class'  # Target column
X = df.drop(columns=[target_col])  # Features
y = df[target_col]  # Target

# Create scaler that maps values to range [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))

# Fit the scaler and transform the data
X_scaled = scaler.fit_transform(X)

print("âœ… Normalization Completed (Range: -1 to 1)")
```

```python
df
```

```python
# Convert the normalized NumPy array back to a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Add the target column back to the DataFrame
normalized_df = pd.concat([X_scaled_df, y.reset_index(drop=True)], axis=1)

# Display the first few rows
print("ðŸ“Š Normalized Data (first 10 rows):")
print(normalized_df.head(10))
```

```python
from sklearn.neural_network import MLPRegressor

hidden_dim = 5
autoencoder = MLPRegressor(hidden_layer_sizes=(hidden_dim,),
                           max_iter=100,
                           random_state=0)
autoencoder.fit(X_scaled_df, X_scaled_df)
```

```python
coefs = autoencoder.coefs_[0]  # weights from input layer to hidden layer
feature_importance = np.sum(np.abs(coefs), axis=1)
```

```python
k = 100
top_features_idx = np.argsort(feature_importance)[::-1][:k]
top_features = X.columns[top_features_idx]

print("Top 20 features:")
print(top_features)
```

```python
# Modified RF training & evaluation (drop-in)
import time
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.utils import check_random_state
import warnings
warnings.filterwarnings("ignore")

RANDOM_STATE = 42

# --- 1) Prepare X and y (flexible: uses existing X_top_scaled / y_top if present) ---
try:
    # if notebook already has X_top_scaled, y_top (numpy or dataframe/series)
    X = X_scaled_df
    y = y
    print("Using existing variables: X_top_scaled and y_top")
except NameError:
    # fallback: try to build from DataFrame df_normalized_randomized and a list top_features
    try:
        df = df_normalized_randomized  # expects a pandas DataFrame already normalized and randomized
        features = top_features        # expects a list of top feature names (strings)
        print("Building X and y from df_normalized_randomized and top_features")
        X = df[features].values
        # assume the target/class column name is 'class' or last column; adjust if needed:
        if 'class' in df.columns:
            y = df['class'].values
        else:
            # if target is last column
            y = df.iloc[:, -1].values
    except Exception as e:
        raise RuntimeError(
            "Could not find X_top_scaled/y_top or df_normalized_randomized/top_features in the notebook namespace.\n"
            "Please set either (X_top_scaled, y_top) or (df_normalized_randomized, top_features) and re-run this cell."
        ) from e

# Convert to numpy arrays (safe)
X = np.array(X)
y = np.array(y)

print(f"X shape: {X.shape}, y shape: {y.shape}")

# --- 2) Quick sanity checks ---
if len(X) != len(y):
    raise ValueError("Number of samples in X and y do not match.")

# Check number of classes and class balance
unique, counts = np.unique(y, return_counts=True)
print("Class distribution:", dict(zip(unique, counts)))

# --- 3) Train/test split (stratified) ---
test_size = 0.40
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, stratify=y, random_state=RANDOM_STATE
)
print(f"Train samples: {len(X_train)}, Test samples: {len(X_test)}")

# --- 4) RandomForest model (simple, reproducible) ---
rf_model = RandomForestClassifier(
    n_estimators=100,   # you can raise this later if needed
    max_depth=None,     # full trees
    random_state=RANDOM_STATE,
    n_jobs=-1
)

start = time.time()
rf_model.fit(X_train, y_train)
fit_time = time.time() - start
print(f"Training done in {fit_time:.2f} sec")

# --- 5) Predictions and evaluations ---
y_pred = rf_model.predict(X_test)
accuracy_test = accuracy_score(y_test, y_pred)
y_pred_train = rf_model.predict(X_train)
accuracy_train = accuracy_score(y_train, y_pred_train)

print(f"\nTrain Accuracy: {accuracy_train:.6f}")
print(f"Test  Accuracy: {accuracy_test:.10f}\n")  # show many decimals as you requested earlier

print("Confusion Matrix (test):")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report (test):")
print(classification_report(y_test, y_pred, digits=6))

# --- 6) Cross-validation (optional quick check) ---
try:
    cv_scores = cross_val_score(rf_model, X, y, cv=5, scoring='accuracy', n_jobs=-1)
    print(f"5-fold CV accuracies: {cv_scores}")
    print(f"CV mean accuracy: {cv_scores.mean():.6f} (+/- {cv_scores.std():.6f})")
except Exception as e:
    print("Could not run cross_val_score:", e)

# --- 7) Feature importances (map back to feature names if available) ---
try:
    importances = rf_model.feature_importances_
    if 'features' in locals() and len(features) == len(importances):
        feat_imp_df = pd.DataFrame({
            "feature": features,
            "importance": importances
        }).sort_values("importance", ascending=False).reset_index(drop=True)
    elif 'X_top_scaled' in globals() and hasattr(X_top_scaled, "columns"):
        feat_imp_df = pd.DataFrame({
            "feature": X_top_scaled.columns,
            "importance": importances
        }).sort_values("importance", ascending=False).reset_index(drop=True)
    else:
        # generic names if no feature names available
        feat_imp_df = pd.DataFrame({
            "feature": [f"f{i}" for i in range(len(importances))],
            "importance": importances
        }).sort_values("importance", ascending=False).reset_index(drop=True)

    print("\nTop feature importances:")
    print(feat_imp_df.head(10).to_string(index=False))
except Exception as e:
    print("Could not compute feature importances:", e)

# --- 8) Helpful checks if accuracy == 1.0 (common leakage/label issues) ---
if np.isclose(accuracy_test, 1.0):
    print("\nWARNING: Test accuracy is 1.0 (or very close). This could indicate:")
    print("- Data leakage (features contain the label or information derived from the label).")
    print("- You accidentally trained and tested on the same samples (check indices before splitting).")
    print("- Very small or trivial dataset resulting in perfect fit.")
    print("Suggestions:")
    print("- Verify your normalized/randomized dataset was shuffled BEFORE splitting.")
    print("- Ensure you didn't include the target column among features.")
    print("- Try a different random_state, or use cross-val to confirm.")
    print("- Print a few rows of X and y and inspect for obvious leaks.")

# --- 9) If you want to save the trained model (optional) ---
# from joblib import dump
# dump(rf_model, "rf_topfeatures_model.joblib")

# End of cell
```
